```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as mplt
```

```python
data = pd.read_csv ('Dataset_heart rate_retail stores.csv')
#print(data)
```

```python
data.head()
```

```python
data.columns = ['cus_Id', 'age', 'gender', 'BMI', 'sleeping_time', 'sporting_activity_h_week', 'daytime', 'outdoor_temperature',
'shopping_cart', 'attendants', 'shopping_frequency', 'cash_point', 'shopping_amount', 'weekday', 'smoking', 'ex_max'
, 'ex_min', 'hr_max', 'max_time', 'hr_min', 'min_time', 'shopping_duration_In_minutes', 'reaching_HRmax']
```

```python
levels = np.linspace(min(data['age']), max(data['age']), 7)
intervals =["18-26", "27-35", "36-44", "45-53", "54-62","63+"]
data["age_groups"]=pd.cut(data["age"], levels, labels=intervals, include_lowest=True)

print(data["age_groups"])
```

```python
mplt.hist(data.age_groups)
mplt.show()
```

```python
levels =np.linspace(min(data['shopping_duration_In_minutes']), max(data['shopping_duration_In_minutes']), 5)
intervals =["06-28", "29-51", "52-74", "75-98"]
data["shopping_dur_min"]=pd.cut(data["shopping_duration_In_minutes"], levels, labels=intervals, include_lowest=True)

print(data["shopping_dur_min"])
```

```python
mplt.hist(data.shopping_dur_min)
mplt.show()
```

### crosstab visualization

```python
# representation of gender on outdoor temperature
pd.crosstab(data['outdoor_temperature'], data['gender'], margins = False).plot.bar();
```

```python
# representation of gender on age groups
pd.crosstab(data['age_groups'], data['gender'], margins = False).iloc[:-1].plot.bar();
```

```python
# representation of age on age smoking
pd.crosstab(data['age_groups'], data['smoking'], margins = False).iloc[:-1].plot.bar();
```

```python
# create shopping duration group
levels = np.linspace(min(data['shopping_duration_In_minutes']), max(data['shopping_duration_In_minutes']), 7)
intervals =["6-21", "21-37", "37-52", "52-67", "67-82","82-98"]
group_shopping_duration = pd.cut(data["shopping_duration_In_minutes"], levels, labels=intervals, include_lowest=True)

# representation of age on shopping duration groups
pd.crosstab(group_shopping_duration, data['age_groups'], margins = False).iloc[:-1].plot.bar();
```
